```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data from the CSV file
data = pd.read_csv('data.csv', sep=';') 

data
```

```python
# Group the data by 'alfa' and plot a line for each group
for name, group in data.groupby('alfa'):
    plt.plot(group['N'], group['latency (s)'], label=f'alfa={name}', marker='o')

# Add labels and title
plt.xlabel('N')
plt.ylabel('Latency (s)')
plt.title('N vs Latency for different alfa values')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
